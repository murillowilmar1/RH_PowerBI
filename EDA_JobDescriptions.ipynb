```python

# Para la distribución por género
gender_counts = df['GenderCode'].value_counts()
gender_counts.plot(kind='bar')
plt.title('Distribución de Empleados por Género')
plt.xlabel('Género')
plt.ylabel('Número de Empleados')
plt.show()
```

```python
import matplotlib.pyplot as plt

gender_counts = df['GenderCode'].value_counts()
labels = gender_counts.index
sizes = gender_counts.values
colors = ['lightcoral', 'lightskyblue']
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

plt.title('GenderCode')
plt.show()
```

### Genero por descripcion de las funciones

```python
contingency_table = pd.crosstab(df['GenderCode'], df['Division'])

# Transponer la tabla de contingencia para gráfico de barras agrupadas
contingency_table = contingency_table.T

# Crear un gráfico de barras agrupadas
contingency_table.plot(kind='bar', figsize=(12, 6))

# Configurar el título y etiquetas
plt.title('Distribución de Division por GenderCode')
plt.xlabel('Division')
plt.ylabel('Número de Empleados')

# Mostrar el gráfico
plt.show()
```

```python
contingency_table = pd.crosstab(df['GenderCode'], df['JobFunctionDescription'])
contingency_table = contingency_table.T
contingency_table.plot(kind='bar', figsize=(20, 10))
plt.title('Distribución de JobFunctionDescription por GenderCode')
plt.xlabel('Division')
plt.ylabel('Número de Empleados')
plt.show()
```

### Calficaciones por empleado segun el genero

```python
female_ratings = df[df['GenderCode'] == 'Female']

# Calcular el promedio de las calificaciones actuales de las mujeres
female_average_rating = female_ratings['Current Employee Rating'].mean()

# Imprimir el promedio
print("Promedio de calificaciones actuales para mujeres:", female_average_rating)
```

```python
male_ratings = df[df['GenderCode'] == 'Male']

# Calcular el promedio de las calificaciones actuales de los Hombres
male_average_rating = male_ratings['Current Employee Rating'].mean()

# Imprimir el promedio
print("Promedio de calificaciones actuales para mujeres:", male_average_rating)
```

```python
gender_rating_means = df.groupby('GenderCode')['Current Employee Rating'].mean()

plt.figure(figsize=(8, 6))
gender_rating_means.plot(kind='bar')
plt.title('Calificación Promedio Actual por GenderCode')
plt.xlabel('GenderCode')
plt.ylabel('Calificación Promedio')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='GenderCode', y='Current Employee Rating', data=df)
plt.title('Distribución de Current Employee Rating por GenderCode')
plt.xlabel('GenderCode')
plt.ylabel('Current Employee Rating')
plt.show()
```
